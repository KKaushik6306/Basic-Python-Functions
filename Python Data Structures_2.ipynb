```python
#Create function which will give sum of even numbers & odd numbers present in some list
lst = [2,3,4,5,6,7,8,9,10]

def evenodd_sum(lst):
    even_sum = 0
    odd_sum = 0
    for i in lst:
        if i%2==0:
            even_sum=even_sum + i
        else:
            odd_sum = odd_sum + i
    return even_sum,odd_sum
```

```python
evenodd_sum(lst)
```

```python
a,b = evenodd_sum(lst)
print('Even sum if',a)
print('Odd sum if',b)
```

# Lambda Funtions 

It is also called Anonymous function. A function with no name. <br>
It can be written only when you have single expression only in function.

```python
def addition(a,b):
    return a+b
```

```python
addition(4,5)
```

```python
addition = lambda a,b:a+b
```

```python
addition(5,6)
```

```python
def even(num):
    if num%2==0:
        print('Yes it is even number')
```

```python
even(2)
```

```python
even = lambda a:a%2==0
```

```python
even(4)
```

```python
addition3 = lambda x,y,z:x+y+z
```

```python
addition3(2,3,4)
```

# List comprehension 

List comprehension provide a concise way to create list. It consists of brackets containing an expression followed by a for clause, then zero or more for or if clause. The expression can be anything, meaning you can put in all kinds of objects in lists.

```python
def lst_squares(lst):
    ind=0
    for i in lst:
        lst[ind] = i**2
        ind+=1
```

```python
a = [2,3,4,5,6,7,8]
lst_squares(a)
a
```

```python
#We can do list comprehension here.
a = [i**2 for i in range(2,9)]
a
```

```python
b = [i**2 for i in a]
b
```

```python
c = [i for i in b if i%2==0]
c
```
